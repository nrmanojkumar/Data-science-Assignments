```python
from sklearn.metrics import accuracy_score

# List of possible K values
k_values = [1, 3, 5, 7, 9, 11, 13, 15]

# Variable to store the maximum accuracy
max_accuracy = 0

# Variable to store the optimal K value
optimal_k = 1

# Iterate over all possible K values to find the optimal one
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    if accuracy > max_accuracy:
        max_accuracy = accuracy
        optimal_k = k

print(f"The optimal K value is {optimal_k} with an accuracy of {max_accuracy}")
```

Visualize the decision boundaries

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

feature1 = 'hair'
feature2 = 'feathers'
```

```python
# Create color maps
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])
```

```python
knn = KNeighborsClassifier(n_neighbors=optimal_k)
knn.fit(X_train[[feature1, feature2]], y_train)
```

```python
mesh_step_size = .01
plot_symbol_size = 50
```

```python
x_min, x_max = X_train[feature1].min() - 1, X_train[feature1].max() + 1
y_min, y_max = X_train[feature2].min() - 1, X_train[feature2].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, mesh_step_size),
                     np.arange(y_min, y_max, mesh_step_size))
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
```

```python
# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)
```

```python
# Plot training points
plt.scatter(X_train[feature1], X_train[feature2], s=plot_symbol_size, c=y_train, cmap=cmap_bold, edgecolor='black')
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())

plt.title(f"2-Class classification (k = {optimal_k}, weights = 'uniform')")
plt.show()
```
