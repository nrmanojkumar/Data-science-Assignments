```python
#1. 1.	Data Cleaning and Preparation:
```

```python
import pandas as pd

df = pd.read_csv('Cardiotocographic.csv')

df.head()
```

```python
df.dtypes
```

```python
# Check for missing values
print(df.isnull().sum())

df.dropna(inplace=True)

for col in df.columns:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python

df['LB'] = pd.to_numeric(df['LB'], errors='coerce')
df['AC'] = pd.to_numeric(df['AC'], errors='coerce')
df['FM'] = pd.to_numeric(df['FM'], errors='coerce')
df['UC'] = pd.to_numeric(df['UC'], errors='coerce')
df['DL'] = pd.to_numeric(df['DL'], errors='coerce')
df['DS'] = pd.to_numeric(df['DS'], errors='coerce')
df['DP'] = pd.to_numeric(df['DP'], errors='coerce')
df['ASTV'] = pd.to_numeric(df['ASTV'], errors='coerce')
df['MSTV'] = pd.to_numeric(df['MSTV'], errors='coerce')
df['ALTV'] = pd.to_numeric(df['ALTV'], errors='coerce')
df['Width'] = pd.to_numeric(df['Width'], errors='coerce')
df['Tendency'] = pd.to_numeric(df['Tendency'], errors='coerce')
df['NSP'] = pd.to_numeric(df['NSP'], errors='coerce')
```

```python
from scipy import stats
import numpy as np

z_scores = np.abs(stats.zscore(df['LB']))
df = df[(z_scores < 3)]
df
```

```python
# Compute the correlation matrix
corr = df.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```
